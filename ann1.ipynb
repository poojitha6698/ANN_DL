```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('train.csv')

print(df.head())
```

```python
# Select the Features and Target variable

# Select the input features (excluding Id and Cover_Type)
x = df.drop(columns=['Id', 'Cover_Type'])

# Target column (subtracting 1 to convert classes 1-7 into 0-6 index ranges for Keras)
y = df['Cover_Type'] - 1
```

x ---> independent variables (54 columns)

y --- > dependent variable (classes 0 through 6)

```python
## Split the dataset into training and testing part

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
## Feature Scaling

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)

x_test = scaler.transform(x_test)
```

```python
# Create the ANN model

model = keras.Sequential([
    # First hidden layer (input_shape adjusted to 54 features)
    keras.layers.Dense(16, activation='relu', input_shape=(54,)),

    # Second Hidden layer
    keras.layers.Dense(8, activation='relu'),

    # Output Layer (7 neurons with softmax activation for multi-class classification)
    keras.layers.Dense(7, activation='softmax')
])
```

```python
### Compile the Model

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model

# Train ANN
history = model.fit(
    x_train,
    y_train,
    epochs=20,
    batch_size=32,
    validation_split=0.2 # use 20% of the split for validation
)
```

```python
# Evaluate the model

loss, accuracy = model.evaluate(x_test, y_test)

print("Accuracy: {:.2f}%".format(accuracy * 100))
```

```python
# Predict

predictions = model.predict(x_test)

# Convert probabilities into the class index with highest probability
predictions = (predictions > 0.5)

print(predictions[:10])
```
