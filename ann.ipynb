```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Churn_Modelling.csv')

print(df.head())
```

```python
# Select the Features and Target variable

#Select the input features 
x = df[["CreditScore","Age","Balance","EstimatedSalary"]]

# Target column
y = df['Exited']
```

x ---> independent variable

y ---> dependent variable(0 or 1)

```python
## Split the dataset into training and testing part

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

80% Training

20% Testing

```python
## Feature Scaling

scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)

x_test = scaler.transform(x_test)
```

##### Build ANN Model

##### Architecture

1. Input Layer

2. Hidden Layer 1

3. Hidden Layer 2

4. Output Layer

```python
# Create the ANN model

model = keras.Sequential([
    # First hidden layer
    keras.layers.Dense(16,activation='relu',input_shape=(4,)),

    # Second Hidden layer
    keras.layers.Dense(8,activation='relu'),

    # Output Layer
    keras.layers.Dense(1,activation='sigmoid')
])
```

##### 16 neurons ---> learn pattern

##### 8 neurons ---> deeper learning

##### Sigmoid ---> binary prediction

```python
### Compile the Model

model.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

##### Adam ---> optimizer

##### Binary crossentropy ---> binary classification loss

##### Accuracy ---> evaluation metric

```python
# Train the model

# Train ANN

history = model.fit(
    x_train,
    y_train,
    epochs = 20,
    batch_size = 32,
    validation_split = 0.2 # use 20% of the split from validation
)
```

##### Epoch ---> one learning cycle

##### Batch size ---> rows proceed together

```python
# Evaluate the model

loss,accuracy = model.evaluate(x_test,y_test)

print("Accuracy: {:.2f}%".format(accuracy*100))
```

```python
# Predict

predictions = model.predict(x_test)

# Convert probabilities into class
predictions = (predictions > 0.5)

print(predictions[:10])
```

False ---> 0

True ---> 1

#### Conclusion : 

If it is zero, he will stay

If it is one, he leaves
